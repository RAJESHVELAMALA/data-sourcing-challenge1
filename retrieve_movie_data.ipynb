### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
import numpy as np
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()

nyt_api_key = os.getenv("NYT_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
```

### Access the New York Times API

```python
# Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
params = {
    'fq': filter_query, 
    'api-key': nyt_api_key,
    'fl':field_list,
    'begin_date':begin_date,
    'end_date':end_date,
    'sort':sort
}
```

```python
# Create an empty list to store the reviews
reviews_list=[]

# loop through pages 0-50
for page in np.arange(0,50):
    # create query with a page number
    # API results show 10 articles at a time
    params['page']=page
    # Make a "GET" request and retrieve the JSON
    response=requests.get(url,params)
    # Add a twelve second interval between queries to stay within API query limits
    time.sleep(12)
    response_json=response.json()
    # Try and save the reviews to the reviews_list
    # loop through the reviews["response"]["docs"] and append each review to the list
    reviews = response_json['response']['docs']
    for review in reviews:
        reviews_list.append(review)
    # Print the page that was just retrieved
    print(f'Checked page {page}')
    # Print the page number that had no results then break from the loop
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
for review_item in reviews_list[:5]:
    print(json.dumps(review_item,indent=4))
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
review_dataframe=pd.json_normalize(reviews_list)
review_dataframe
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early

unicode_char1 = '\u2018'
unicode_char2 = '\u2019 '
unicode_char3 = '\u2019'

def splitStr(text):
    review_txt=str(text)
    title= review_txt
    try:
        start_index=review_txt.index(unicode_char1)
        try:
            end_index=review_txt.index(unicode_char2)  
        except Exception as e:
            end_index=review_txt.index(unicode_char3)  
        title = review_txt[start_index+1:end_index]
    except Exception as e:
        title=review_txt

    if len(title)<=0: title=review_txt
    return title

review_dataframe['title']=review_dataframe['headline.main'].apply(splitStr)

review_dataframe
```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords

# Fix the "keywords" column by converting cells from a list to a string
review_dataframe['keywords']=review_dataframe['keywords'].apply(extract_keywords)

review_dataframe
```

```python
# Create a list from the "title" column using to_list()
# These titles will be used in the query for The Movie Database

title_list=review_dataframe['title'].to_list()
title_list
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query=The Attachment Diaries"
movie_url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key

movie_detail_url='https://api.themoviedb.org/3/movie/'
apikey = "api_key="+ tmdb_api_key

#finalurl = url+tmdb_key_string
#response= requests.get(finalurl)
#responseJson=response.json()
```

```python
def movieDeails(responseJson):
    for movie in responseJson['results']:
        # Get movie id
        movieid=movie['id']
        # Make a request for a the full movie details
        movie_url_final=movie_detail_url+str(movieid)+"?"+apikey
        moviedetails=requests.get(movie_url_final)
        movieJson=moviedetails.json()
        # Extract the genre names into a list
        genre_dict=movieJson['genres']
        genre_names = [genre['name'] for genre in genre_dict]
        # Extract the spoken_languages' English name into a list       
        spoken_languages=movieJson['spoken_languages']       
        languages = [lang['english_name'] for lang in spoken_languages]
        # Extract the production_countries' name into a list
        production_countries=movieJson['production_countries']
        countries = [countries['name'] for countries in production_countries]
        # Add the relevant data to a dictionary and
        movieJson['spoken_languages']=languages
        movieJson['production_countries']=countries
        movieJson['genres']=genre_names
        # append it to the tmdb_movies_list list
        tmdb_movies_list.append(movieJson)
        #print(json.dumps(movieJson,indent=4))
```

```python
# Create an empty list to store the results
tmdb_movies_list=[]

# Create a request counter to sleep the requests after a multiple
# of 50 requests
req_counter = 0
# Loop through the titles
for title in title_list:
    # Check if we need to sleep before making a request
    if(req_counter==50):
        req_counter=0
        time.sleep(15)
    # Add 1 to the request counter
    req_counter+=1
    #print('-'*50)
    #print(title)
    try:
        movieurl = movie_url+title+tmdb_key_string
        print(f'Movie # {title} found')
        movie_response=requests.get(movieurl)
        movie_responsejson = movie_response.json()
        movieDeails(movie_responsejson)
    except Exception as e:
        print(f'Movie {title} not found')
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data

tmdb_movies_list
```

```python
# Convert the results to a DataFrame
movie_list_df=pd.DataFrame(tmdb_movies_list)
movie_list_df
```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title

mergeddf=review_dataframe.merge(movie_list_df,on='title')
mergeddf
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
listColumns=['production_countries','spoken_languages']

# Create a list of characters to remove
chars_remove =['[',']','\'']

# Loop through the list of columns to fix
for col in listColumns:
    # Convert the column to type 'str'
    mergeddf[col] = mergeddf[col].astype(str)
    # Loop through characters to remove
    for chars in chars_remove:
        mergeddf[col] = mergeddf[col].str.replace(chars,'')

# Display the fixed DataFrame
mergeddf
```

```python
# Drop "byline.person" column
mergeddf.columns

mergeddf.drop(columns=['byline.person'],inplace=True)

mergeddf.columns
```
